```python
#I hate this fix so much

def time_formatter(time_string):
    time_string = time_string.removesuffix('s')
    if len(time_string) == 1:
        time_string = '00:00:0' + time_string
    elif len(time_string) == 2:
        time_string = '00:00:' + time_string
    elif len(time_string) == 3:
        time_string = '00:00' + time_string
    elif len(time_string) == 4:
        time_string = '00:0' + time_string
    elif len(time_string) == 5:
        time_string = '00:' + time_string
    elif len(time_string) == 6:
        time_string = '00' + time_string
    return time_string
    
```
