### Loop for counting the number of training sessions

```python
for k in range(0, len(race_data)):
    try: #count the number of training sessions
        id = race_data.loc[k, 'strava_id']
        race_data.loc[k, 'no_sessions'] = \
            len(training_data[training_data['strava_id'] == id])
    except KeyError: #keep 0 as the value
        race_data.loc[k, 'no_sessions'] = 0
```

```python
len(race_data[race_data['no_sessions'] >= 10])
```

```python
len(race_data)
```

```python
#keep only those athletes with at least 10 training sessions?
race_data = race_data[race_data['no_sessions'] >= 10]
```

```python
# reset the index now that columns have been dropped
race_data = race_data.reset_index().drop(columns=['index'])
```

```python
race_data
```

### Loop for adding the mean training distance to the race data

```python
mean_train_dist = training_data[['strava_id', 'distance']].groupby(by='strava_id').mean()

for k in range(0, len(race_data)):
    id = race_data.loc[k, 'strava_id']
    race_data.loc[k, 'mean_train_dist'] = \
        mean_train_dist.loc[id,'distance']
```

```python
race_data
```

### Loop for computing mean average speed

```python
training_data.head()
```

```python
#compute average speed for a given ride and convert to km/h for interpretability
training_data['average_speed'] = training_data.distance/training_data.time * 3600

mean_train_avg_speed = training_data[['strava_id', 'average_speed']].groupby(by='strava_id').mean()

for k in range(0, len(race_data)):
    id = race_data.loc[k, 'strava_id']
    race_data.loc[k, 'mean_train_avg_speed'] = mean_train_avg_speed.loc[id, 'average_speed']
        
```
